## Valores do seguro - PROAGRO
RISCO - É o Limite Máximo Indenizável. Calculado como Valor do Orçamento de Custeio (R$/área) x área. O Orçamento de Custeio depende da produtividade esperada e do preço do produto.

ADICIONAL - É o valor pago pelo empreendedor.

SINISTRO - Despesa total com indenização que inclui: Valor Financiado + Recursos Próprios + Parcela de Investimentos + Garantia de Renda Mínima (GRM). 

Z - É a razão do SINISTRO por RISCO.



```python
dados.groupby(['N_EVENTO']).aggregate({'R$ ADICIONAL': [np.mean,np.sum, 'count'], 'R$ SINISTRO': [np.mean,np.sum], 'R$_RISCO': [np.mean,np.sum], 'Z': np.mean, 'AREA_ha': np.mean}).sort_values(by = ('R$ ADICIONAL', 'count'), ascending = False)
```

Aparentemente, o valor do risco varia bastante com o tipo de sinistro. Por exemplo, o RISCO médio para eventos do tipo SECA, estão em torno dos 14k, enquanto que para sinistros do tipo GEADA, o valor ultrapassa os 40k. O alto valor do RISCO para os casos de geada se justifica porque os 4 casos de geada aconteceram em lavouras com grande área média.

Por outro lado, o valor do SINISTRO é definido em função das perdas do agricultor após o acontecimento do sinistro. É razoável pensar que a razão Z (SINISTRO/RISCO) é um bom indicador da destrutividade do evento extremo. Nesse caso, GEADA, CHUVA EXCESSIVA e SECA seriam os eventos mais destrutivos e VARIAÇÃO EXCESSIVA DE TEMPERATURA os menos destrutivos. 



#### Seguindo esse raciocínio, os municípios com valor de Z mais alto na tabela abaixo provavelmente foram atingidos por eventos mais destrutivos.

```python
dados.groupby(['UF-MINICIPIO']).aggregate({'Z': ['count', np.mean], 'R$ SINISTRO': np.mean, 'R$_RISCO':np.mean}).sort_values(by = ('Z', 'mean'), ascending = False)
```

### Orçamento de Custeio
Outra variável importante que não está descrita no nosso dataset é o Orçamento de Custeio (Produção Esperada [R$] / Área [ha]). Ela reflete a avaliação que o segurador faz da lavoura do produtor. Podemos adicioná-la ao dataset fazendo ORCAMENTO = RISCO/AREA

```python
dados['ORCAM_CUSTEIO']= dados['R$_RISCO']/dados['AREA_ha']
```

### Hipótese:
O orçamento de custeio deve ser similar para duas lavouras de um mesmo município e cultura. Então, espera-se que o desvio padrão relativo (rsd) seja pequeno, numa mesma cidade para uma mesma cultura.

```python
print(dados['ORCAM_CUSTEIO'].std())
result = dados.groupby(['UF-MINICIPIO', 'N_SICOR']).aggregate({'ORCAM_CUSTEIO': [np.mean, np.std, 'count']})
result['rsd'] = 100*result[('ORCAM_CUSTEIO', 'std')]/result[('ORCAM_CUSTEIO', 'mean')]
print(result)
```

#### Alguns valores de desvio médio relativo (rsd) do orçamento por hectare ainda é alto mesmo em grupos de município e cultura semelhantes. Uma possível explicação para a discrepância é a diferença do maquinário para dois agricultores
