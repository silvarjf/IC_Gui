## Valores do seguro - PROAGRO
RISCO - É o Limite Máximo Indenizável. Calculado como Valor do Orçamento de Custeio (R$/área) x área. O Orçamento de Custeio depende da produtividade esperada e do preço do produto.

ADICIONAL - É o valor pago pelo empreendedor.

SINISTRO - Despesa total com indenização que inclui: Valor Financiado + Recursos Próprios + Parcela de Investimentos + Garantia de Renda Mínima (GRM). 

Z - É a razão do SINISTRO por RISCO.



```python
dados.groupby(['N_EVENTO']).aggregate({'R$ ADICIONAL': [np.mean,np.sum, 'count'], 'R$ SINISTRO': [np.mean,np.sum], 'R$_RISCO': [np.mean,np.sum], 'Z': np.mean, 'AREA_ha': np.mean}).sort_values(by = ('R$ ADICIONAL', 'count'), ascending = False)
```

- Por que a variação nos valores? Geada e Doença risco são  mais altos. 
    Diferença nos preços e níveis de produtividade

Como é calculado o orçamento de custeio? Produtivadade por hectare vezes preço da saca
- A produtividade pode variar mesmo entre produtores da mesma região e espécie plantada, logo a relação entre orçamento de custeios e essas categorias não é constante.

```python
dados['ORCAM_CUSTEIO']= dados['R$_RISCO']/dados['AREA_ha']
```

```python
print(dados['ORCAM_CUSTEIO'].std())
result = dados.groupby(['UF-MINICIPIO', 'N_SICOR', 'N_EVENTO' ]).aggregate({'ORCAM_CUSTEIO': [np.mean, np.std, 'count']})
result['rsd'] = 100*result[('ORCAM_CUSTEIO', 'std')]/result[('ORCAM_CUSTEIO', 'mean')]
print(result)
```

```python
O desvio médio relativo (rsd) do orçamento por hectare ainda é alto mesmo em grupos de município, espécie plantada e tipo de sinistro semelhantes.
Não é possível estabelecer uma relação com o alto rsd e as categorias analisadas (município, espécie plantada e tipo de sinistro)
Hipóteses:
    - O orçamento da terra leva em consideração o maquinário do agricultor
```
